```python
import pandas as pd
data = {
         "compound" : ["Fe2O3", "TiO2", "ZnO"],
         "band_gap" : [2.2, 3.2, 2.4],
         "density" : [5.24, 4.23, 5.61],
}

df = pd.DataFrame(data)
print(df[["compound", "band_gap"]])
```

```python
high_gap = df[df["band_gap"] > 3.0]
print(high_gap)
```
