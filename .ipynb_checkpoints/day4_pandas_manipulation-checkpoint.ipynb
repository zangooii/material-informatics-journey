```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target
avg_by_species = df.groupby("species").mean()
print(avg_by_species)
```

```python
print(df.groupby("species").max())
print(df.groupby("species").count())
print(df.groupby("petal width (cm)").mean())
```

```python
df["sepal_ratio"] = df["sepal length (cm)"] / df["sepal width (cm)"]
print(df.head())
```

```python
compounds = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Fe2O3", "TiO2", "ZnO"]
})

properties = pd.DataFrame({
    "id": [1, 2, 3],
     "am" : [7, 8 ,None],
    "band_gap": [2.2, 3.2, 3.4]
})
merged = pd.merge(compounds, properties)
print(merged)
```

```python
avg = merged.groupby("name").mean()
print(avg)
```

```python
merged["idband_gap"] = merged["id"] + merged["band_gap"]
print(merged)
```

```python
print(merged.dropna())
```

```python
print(merged.dropna(subset=["band_gap"]))
```

```python
print(merged.fillna(0))
```

```python
print(merged.fillna(merged.mean(numeric_only = True)))
```

```python
merged.dropna(inplace = True)
print(merged)
```
