```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [10, 20, 30, 40])
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [4, 8, 12, 16]
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("my second line plot")
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=5)
plt.title("histogram example")
plt.show()
```

```python
categories = ["Fe2O3", "TiO2", "ZnO"]
values = [2.2, 3.2, 3.4]
plt.bar(categories, values)
plt.xlabel("Compound")
plt.ylabel("Band gap (eV)")
plt.title("Band gap comparison")
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df["species"] = iris.target
df["sepal length (cm)"].hist()
plt.title("sepal length histogram")
plt.show()
```

```python
avg_by_species = df.groupby("species").mean()
plt.bar(avg_by_species.index, avg_by_species["sepal length (cm)"])
plt.show()
```

```python
print(avg_by_species)
print(avg_by_species.index)
```
