```python
numList = [1,12,19,12,17,4,2]
def average (y):
    summ = 0
    count = 0
    for x in y:
        summ = summ + x
        count = count + 1
    ave = summ / count
    return ave
result = average(numList)
print(result)
```

```python
ourList = ["snack", "sandwich", "ear", "flower", "flower"]
def howManyTimes(ourList, var):
    counter = 0
    for x in ourList:
        if x == var:
            counter = counter + 1
    return counter
result = howManyTimes(ourList, "flower")
print(result)
```

```python
def isPrime(x):
    if x > 1 and x % 1 ==0:
        half = x/2
        for y in range(half):
            if x % y == 0 and x/y != 1 and x:
                return "Prime"
            else return "notPrime"    
result = isPrime(23)
print(result)
```

```python
def isPrime(x):
    if x % 1 == 0 and x > 1:
        half = x//2
        for y in range(2, half + 1):
            if x % y == 0 :
                return "notPrime"
            else:
                return "prime"
    return "notPrime"    
result = isPrime(25)
print(result)
```

```python
numList = [25,8,9,10,11,12,13]
def min_max(numList):
    min = numList[0]
    max = numList[0]
    for x in numList:
        if x > max:
            max = x
        if x < min:
            min = x
    return (min, max)
result = min_max(numList)
print(result)
```

```python
numList = [25,8,9,10,11,12,13]
def min_max(numbers):
    return min(numbers), max(numbers)
result = min_max(numList)
print(result)
```

```python
def reverse(s):
    result = ""
    for l in s:
        result = l + result
    print(result)
reverse("materialInformatics")
```

```python
def wordCount(sentence):
    count = {}
    words = sentence.split()
    for word in words:
        if word in count:
            count[word] = count[word] + 1
        else:
            count[word] = 1
    return count
result = wordCount("I am getting crazy with this codes I want to get to the exciting parts")
print(result)
```
