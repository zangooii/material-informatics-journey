```python
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()
df = pd.DataFrame(wine.data, columns= wine.feature_names)
df["target"] = wine.target

print(df.describe())
```

```python
print(df.isnull())
```

```python
print("max")
print(df.groupby("target").max())

print("min------------------------------------------------------------------------------------------------------------")
print(df.groupby("target").min())
```

```python
print(df.groupby("target").count())
```

```python
print(df.groupby("alcohol").mean())
```

```python
import matplotlib.pyplot as plt
print(df.groupby("target").mean())
```

```python
by_malic_acid = df.groupby("malic_acid").mean()
plt.plot(by_malic_acid.index, by_malic_acid["ash"])
plt.xlabel("Malic acid")
plt.ylabel("Average ash")
plt.title("Average ash by malic acid level")
plt.show()
```

```python
print(df.groupby("malic_acid").mean())
```

```python
by_class = df.groupby("target").mean()
plt.bar(by_class.index, by_class["flavanoids"])
plt.xlabel("Wine class")
plt.ylabel("Average flavanoids")
plt.title("Average flavanoids by wine class")
plt.show()
```

```python
df["alcohol"].hist(bins=100)
plt.title("alcohol histogram")
plt.show()
```

```python
target = df.groupby("target").mean()
plt.bar(target.index, target["alcohol"])
plt.title("based on alcohol")
plt.show()

plt.bar(target.index, target["ash"])
plt.title("based on ash")
plt.show()

plt.bar(target.index, target["magnesium"])
plt.title("based on magnesium")
plt.show()

plt.bar(target.index, target["color_intensity"])
plt.title("based on color_intensity")
plt.show()

plt.bar(target.index, target["malic_acid"])
plt.title("based on malic_acid")
plt.show()

plt.bar(target.index, target["total_phenols"])
plt.title("based on total_phenols")
plt.show()
```

```python
by_color_intensity = df.groupby("color_intensity").mean()
plt.scatter(by_color_intensity.index, by_color_intensity["malic_acid"])
plt.show()
plt.plot(by_color_intensity.index, by_color_intensity["malic_acid"])
plt.show()
```

This dataset contains chemical measurements from 3 different wine classes. Magnesium, ash, and alcohol showed relatively tight distributions across samples — most alcohol values fell between 12% and 13.2%. Color intensity and malic acid varied more widely. Initially, I suspected a relationship between color intensity and malic acid, but scatter and line plots showed no visible correlation between them.
