```python
test_list = ["Fe", "Cu", "Fe", "Zn", "Fe", "Cu", "Ni"]
def how_many(l):
    count = {}
    for x in l:
        if x in count:
            count[x] = count[x] + 1
        else:
            count[x] = 1
    return count
result = how_many(test_list)
print(result)
```

```python
import pandas as pd
from sklearn.datasets import load_wine

wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df["target"] = wine.target
print(df.isnull().sum())
print(df.groupby("target").mean())
```

```python
import matplotlib.pyplot as plt
average = df.groupby("target").mean()
plt.bar(average.index, average["hue"])
plt.show()
```
